```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os
import json
# this will be disabled for privacy purpose
from config import g_keys
```

```python
# read in the weather dataframe 
# used relative path
vacation_df = pd.read_csv('../weatherPy/weather_df')
vacation_df.head()

# creating dataframe with humidity and cities
locations1 = vacation_df[['Latitude', 'Longitude']]
locations1

# make sure humidity data is float
humidity_data = vacation_df['Humidity'].astype(float)
humidity_max = humidity_data.max()
humidity_max
```

```python
# plot humidity heat maps
fig = gmaps.figure()

heat_maps1=gmaps.heatmap_layer(locations1, 
                               weights=humidity_data, 
                               max_intensity=humidity_max, 
                               point_radius=2, 
                               dissipating=False)

fig.add_layer(heat_maps1)

fig
```

```python
# narrow dataframe according to requirements
# req.1 temp 70>x<80
# req.2 windspeed <10mph
# req.3 cloudiness = 0

temp_df=vacation_df[(vacation_df['Temperatures']>70)&(vacation_df['Temperatures']<80)]
#temp_df.head()
windspeed_df=temp_df[temp_df['Windspeed']<10]
#windspeed_df
final_df=windspeed_df[windspeed_df['Cloudiness']==0]
final_df.reset_index()
#print(len(final_df))
#final_df
```

```python
# generate lists of items required
hotel_name=[]
hotel_address=[]
hotel_latitude=[]
hotel_longitude=[]
hotel_rating=[]

print('---------------\nHotel information\n---------------')
# iterate through each row
for index, row in final_df.iterrows():
    city = row['City']
    coordinates=f"{row['Latitude']},{row['Longitude']}"
    keyword='lodging'
    
    # set params
    params={
        'location':coordinates,
        'type':'hotel',
        'keyword':keyword,
        'radius':5000,
        'key':g_keys
    }
    
    base_url='https://maps.googleapis.com/maps/api/place/nearbysearch/json'
    response=requests.get(base_url,params=params).json()
    # Check the url to indentify appropriate keys and value to parse in
    # print(json.dumps(response, indent=4, sort_keys=True))
    try:
        hotel_name.append(response['results'][0]['name'])
        hotel_address.append(response['results'][0]['vicinity'])
        hotel_latitude.append(response['results'][0]['geometry']['location']['lat'])
        hotel_longitude.append(response['results'][0]['geometry']['location']['lng'])
        hotel_rating.append(response['results'][0]['rating'])
        print(f'processing info for {index} in {city}')
        # add to the df
        final_df.loc[index, 'Hotel Name']=response['results'][0]['name']
    except:
        print(f'no information for this lodging. Skipping...')
print('------------------\nEND')
```

```python
final_df.dropna()
final_df
```

```python
# plot layered maps using hotel coordinates
# get locations first
locations=final_df[['Latitude','Longitude']]

# create info box 
info_box_build="""
<d1>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
</d1>
"""
info_box=[info_box_build.format(**row)for index, row in final_df.iterrows()]

# add a layer to heat maps
marker_layer=gmaps.marker_layer(locations, info_box_content=info_box)
fig.add_layer(marker_layer)
fig
```
